Compile the data from WRA (2017 data are updated) to list how many data including time series are in each well.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['savefig.bbox'] = 'tight'

import datetime
date = datetime.datetime.now().strftime('%Y%m%d')

%matplotlib inline
```

# Read 濁水溪沖積扇水質資料1992-2019.xlsx
The data from 2017 are updated.

```python
wa_dict = pd.read_excel('data/濁水溪沖積扇水質資料1992-2019.xlsx', sheet_name=None)
len(wa_dict)
```

```python
wa_dict.keys()
```

```python
sheets = ['2019', '2018', '2017', '2016', '2015', '2013', '2012', '2011', '2010', '2009', '2008', '2007', '2006', '2002', '2001', '2000', '1999', '1997', '1996', '1994', '1993', '1992']
```

```python
wa_df = pd.DataFrame()

for sheet in sheets:
    wa_ind = pd.read_excel('data/濁水溪沖積扇水質資料1992-2019.xlsx', sheet_name=sheet)
    wa_ind['sheet'] = [sheet for _ in range(len(wa_ind))]
    wa_df = pd.concat([wa_df, wa_ind], axis=0, join='outer')
```

```python
wa_df
```

# Drop the columns and rows not having any value

```python
wa_df.shape
```

```python
wa_df.dropna(axis=1, how='all', inplace=True)
wa_df.shape
```

```python
wa_df.dropna(axis=0, how='all').shape
```

There is no empty raw, but 2 empty columns.

# Assign id

```python
sdate = wa_df['取樣日期'].astype(str)
wa_df['id'] = ['{}_{}'.format(i, j) for i, j in zip(wa_df['井號'], sdate)]
```

```python
wa_df.set_index('id', inplace=True)
wa_df
```

# Check duplicates

```python
wa_du = wa_df[wa_df.index.duplicated()]
wa_du.shape
```
